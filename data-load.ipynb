```python
src_id_col = 'node1'
tgt_id_col = 'node2'
src_label_col = 'node1;label'
tgt_label_col = 'node2;label'

print(f"Source ID Column:     {src_id_col}")
print(f"Target ID Column:     {tgt_id_col}")
print(f"Source Label Column:  {src_label_col}")
print(f"Target Label Column:  {tgt_label_col}")
```

```python
def safe_str(x):
    """Safely converts any value to a Dgraph-compatible string."""
    if pd.isna(x):
        return ""
    return str(x).replace('\\', '\\\\').replace('"', '\\"')


def upsert_edge(txn, sid, slabel, oid, olabel):
    """
    Adds operations to a transaction to create/update two nodes and link them.
    """
    
    query = """
    query q($sid: string, $oid: string) {
      s as var(func: eq(id, $sid))
      o as var(func: eq(id, $oid))
    }
    """
    
    nquads = f'''
      uid(s) <id> "{safe_str(sid)}" .
      uid(s) <label> "{safe_str(slabel)}" .
      uid(s) <dgraph.type> "Node" .

      uid(o) <id> "{safe_str(oid)}" .
      uid(o) <label> "{safe_str(olabel)}" .
      uid(o) <dgraph.type> "Node" .

      uid(s) <rel> uid(o) .
    '''

    mu = pydgraph.Mutation(set_nquads=nquads.encode('utf-8'))
    
    req = pydgraph.Request(
        query=query,
        mutations=[mu],
        vars={"$sid": safe_str(sid), "$oid": safe_str(oid)},
    )
    return txn.do_request(req)


def load_edges_from_df(df_to_load, batch_size=500):
    """
    Iterates over the DataFrame and loads data in batches.
    (v3: Fixes 'StartTs mismatch' error with better txn handling)
    """
    total_rows = len(df_to_load)
    print(f"Starting to load {total_rows} edges in batches of {batch_size}...")
    
    txn = client.txn() 
    mutations_in_batch = 0
    
    try:
        for index, row in df_to_load.iterrows():
            sid = row[src_id_col]
            oid = row[tgt_id_col]
            slabel = row[src_label_col] if pd.notna(row[src_label_col]) else sid
            olabel = row[tgt_label_col] if pd.notna(row[tgt_label_col]) else oid

            # Skip rows with missing IDs
            if not sid or not oid or pd.isna(sid) or pd.isna(oid):
                continue
                
            # Add the upsert operation to the transaction
            upsert_edge(txn, sid, slabel, oid, olabel)
            mutations_in_batch += 1
            
            # If we hit the batch size, commit the transaction
            if mutations_in_batch >= batch_size:
                txn.commit()
                print(f"Committed batch. Processed {index + 1}/{total_rows} rows.")
                txn = None # Mark as committed
                txn = client.txn() # Start a new transaction
                mutations_in_batch = 0
        
        # Commit any remaining mutations after the loop finishes
        if mutations_in_batch > 0:
            txn.commit()
            print(f"Committed final batch. Total processed: {total_rows} rows.")
            txn = None # Mark as committed
            
    except Exception as e:
        print(f"An error occurred: {e}")
        print("Discarding transaction.")
        if txn:
            txn.discard()
            txn = None
    finally:
        if txn:
            txn.discard()
```
