```python
import pydgraph

client_stub = pydgraph.DgraphClientStub('localhost:9080')
client = pydgraph.DgraphClient(client_stub)
```

### Load schema

```python
import os

schema_path = os.path.join(os.getcwd(), '@schema.dql')
schema = ""
with open(schema_path, 'r') as f:
    schema = f.read()

op = pydgraph.Operation(schema=schema, run_in_background=True)
client.alter(op)
```

### Insert

```python
def insert(obj):
    txn = client.txn()
    try:
        response = txn.mutate(set_obj=obj, commit_now=True)
        print(response.uids)
    finally:
        txn.discard()
```

```python
person = {
    'id': 'test',
    'label': 'test'
}

insert(person)
```
